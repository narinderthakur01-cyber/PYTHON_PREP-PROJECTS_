```
import time

n = 10_000_000

a = [1] * n
b = [1] * n

start = time.time()

c = [0] * n
for i in range(n):
    c[i] = a[i] + b[i]

end = time.time()

print("Python List Time:", end - start, "seconds")
```

```
import numpy as np
import time

n = 10_000_000

start = time.time()

a = np.arange(n)
b = np.arange(n)
c = a + b

end = time.time()

print("NumPy Time:", end - start, "seconds")
```
