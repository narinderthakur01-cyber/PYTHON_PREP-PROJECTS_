```
import numpy as np
a=np.array([[1,2,3],[0,1,0],[0,0,1]])
b=np.array([[0,1,0],[1,0,0],[1,2,1]])
# A and B matrix
```

```
a.shape
```

```
print(a+b)
#addition
```

```
#diff
print(a-b)
```

```
#product1
print(a@b)
#product 2
print(np.matmul(a,b))
```

```
#transpose1
print(a.T)
#transpose2
print(np.transpose(a))
```

```
#determinant
print(np.linalg.det(a))
print(np.linalg.det(b))
```

```
#finding Inverse
print(np.linalg.inv(a))
```

```
#finding trace
print(np.linalg.trace(a))
print(np.linalg.trace(b))
```

```
#rank
print(np.linalg.matrix_rank(a))
print(np.linalg.matrix_rank(b))
```

```
#finding eigen value and eigen vector
print(np.linalg.eig(a))
print(np.linalg.eig(b))
```
